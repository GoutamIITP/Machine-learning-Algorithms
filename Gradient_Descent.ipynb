```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate data
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
X = X.flatten()  # Flatten X to a 1D array if necessary

# Gradient descent function with plotting
def gradient_descent(X, y):
    m_curr = b_curr = 0  # Initial values for slope and intercept
    rate = 0.01          # Learning rate
    n = len(X)           # Number of samples

    # Plot original data points
    plt.figure(figsize=(8, 6))
    plt.scatter(X, y, color='red', marker='+', linewidth=5)

    # Gradient descent iterations
    for i in range(10000):
        y_predicted = m_curr * X + b_curr
        if i % 500 == 0:  # Plot every 500 iterations for clarity
            plt.plot(X, y_predicted, color='green', alpha=0.3)

        # Calculate gradients
        md = -(2/n) * sum(X * (y - y_predicted))
        yd = -(2/n) * sum(y - y_predicted)

        # Update parameters
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd

    # Final line of best fit
    plt.plot(X, m_curr * X + b_curr, color='blue')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Gradient Descent Linear Regression")
    plt.show()
    return m_curr, b_curr
# Run the gradient descent function
gradient_descent(X, y)
```

```python
import numpy as np

# Assuming m_curr and b_curr are the final values obtained from gradient descent
# and X, y are the data points
def calculate_mse(X, y, m, b):
    n = len(X)
    y_predicted = m * X + b
    mse = (1/n) * sum((y - y_predicted) ** 2)
    return mse

# Use the values obtained from gradient descent
final_m = 78.3506366794738  # Final slope from gradient descent
final_b = 26.15963284313284  # Final intercept from gradient descent

# Calculate the MSE with the final values
mse = calculate_mse(X, y, final_m, final_b)
print("Mean Squared Error:", mse)
```
