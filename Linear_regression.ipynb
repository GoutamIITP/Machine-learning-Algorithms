The goal of linear regression is to find the best-fitting line through data points by minimizing the sum of squared errors (SSE) between predicted and actual values. This method, known as Ordinary Least Squares (OLS), ensures that the predicted line captures the trend in data with minimal error.

Linear regression is widely used in applications such as forecasting, risk assessment, and trend analysis due to its simplicity and interpretability. However, it assumes linearity, homoscedasticity (constant variance of errors), independence of observations, and normally distributed errors, which may not always hold in real-world scenarios.

Despite its limitations, linear regression remains a foundational tool in predictive analytics and machine learning, serving as a benchmark model and a stepping stone to more complex methods.

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

By using Manual code

```python
#siple linear regression
from sklearn.metrics import mean_squared_error
class MY_LR:
  def __init__(self):
    self.m = None
    self.b = None

  def fit(self, X_train, y_train):
    num = 0
    den = 0

    for i in range(X_train.shape[0]):
      num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
      den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))

    self.m = num/den
    self.b = y_train.mean() - (self.m * X_train.mean())
    print(self.m)
    print(self.b)

  def predict(self, X_test):
    return self.m * X_test + self.b

  def evaluate(self, X_test, y_test):
    y_pred = self.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error: {mse}")
    return mse


```

```python
model = MY_LR()
model.fit(X_train,y_train)
predict = model.predict(X_test)
model.evaluate(X_test,y_test)

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R¬≤): {r2}")
```
