The coefficients are estimated by minimizing the cost function, typically using Mean Squared Error (MSE). Multiple regression uses methods like gradient descent or normal equations to find the optimal values for these coefficients, aiming to reduce the residuals or errors between predicted and actual values.

This type of regression is particularly useful in scenarios where a single predictor isn‚Äôt sufficient to make accurate predictions, and several variables collectively influence the outcome. For example, in predicting house prices, factors like location, number of bedrooms, square footage, and age of the house might all be relevant.

Assumptions in multiple regression include linearity (a linear relationship between each predictor and the target), independence of observations, homoscedasticity (constant variance of residuals), and no multicollinearity (independent variables should not be highly correlated with each other).

Multiple regression helps reveal the individual impact of each variable while controlling for others, allowing us to understand complex relationships. The model‚Äôs performance can be evaluated using metrics like R-squared and Adjusted R-squared, which indicate how well the predictors explain the variance in the target variable.




```python
import numpy as np
from sklearn.datasets import load_diabetes
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

By using manual code algorithm implementation

```python
class My_LR:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self,X_train,y_train):
        X_train = np.insert(X_train,0,1,axis=1)

        # calculate coefficients using normal equations
        betas = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
        self.intercept_ = betas[0]
        self.coef_ = betas[1:]


    def predict(self,X_test):
        y_pred = np.dot(X_test,self.coef_) + self.intercept_
        return y_pred
```

```python
model = My_LR()
model.fit(X_train,y_train)

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R¬≤): {r2}")
```

By Using skLearn library

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R¬≤): {r2}")
```
