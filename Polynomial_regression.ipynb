# Theory

 Polynomial Regression is a regression technique that models the relationship between a
dependent variable and one or more independent variables using a polynomial function. It
extends the concept of Linear Regression by allowing for curved relationships between the
variables

# Polynomial regression work

**Polynomial Features**: Polynomial Regression involves transforming the original features into
polynomial features by raising them to different powers. For example, if the original feature is x,
polynomial features could include x^2, x^3, and so on.

**Degree of Polynomial**: The degree of the polynomial determines the complexity of the model. A
higher degree allows for more flexibility in fitting the data but can also lead to overfitting.

**Model Equation**: The general equation for Polynomial Regression is:

 y = b0 + b1x + b2x^2 + ... + bn*x^n

 where:

 y is the dependent variable

 x is the independent variable

 b0, b1, b2, ..., bn are the coefficients

 n is the degree of the polynomial

# Applications:

 * Data with Non-linear Relationships: Polynomial regression is used when data shows a non-linear trend that cannot be captured by a simple linear model, such as in economic modeling, population growth trends, and engineering applications.

 * Predictive Modeling: It provides a more flexible model compared to linear regression, allowing for better prediction when the relationship between variables is inherently non-linear.

Using Library


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate sample data
np.random.seed(0)
X = 6 * np.random.rand(100, 1) - 3  # Random values in the range [-3, 3]
y = 0.5 * X**3 + X**2 + 2 + np.random.randn(100, 1) * 3  # Cubic relationship with noise

# Visualize the data
plt.scatter(X, y, color='blue', label='Data Points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sample Data')
plt.legend()
plt.show()
```

```python
# Transform the data to include polynomial features
degree = 3  # Degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Fit the Linear Regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict using the model
X_new = np.linspace(-3, 3, 100).reshape(-1, 1)  # New data for predictions
X_new_poly = poly_features.transform(X_new)
y_pred = model.predict(X_new_poly)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X_new, y_pred, color='red', label='Polynomial Regression Model', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()

# Print the coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
